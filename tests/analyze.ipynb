```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
```

```python
# Reset display options to the default values if needed
# Set display options to show more rows and columns
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
```

```python
file1 = 'proglat_2022.xlsx'
file2 = 'proglat_2023.xlsx'

df1 = pd.read_excel(file1)
df1.head()
```

```python
len(df1)
```

```python
df2 = pd.read_excel(file2)
df2.head()
```

```python
len(df2)
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df.head()
```

```python
len(df)
```

```python
# Check duplicated data
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df.columns
```

```python
selectedCols =['instansi_klas', 'instansi', 'program', 'kejuruan', 'Subkejuruan', 'jenjangKKNI', 'levelKBJI' ]
df_selected = df[selectedCols]
df_selected.head()
```

```python
df_selected.tail()
```

```python
for i in selectedCols:
    print(i, '\t:', len(df_selected.value_counts(i)))
```

```python
df_selected.value_counts('levelKBJI')
```

```python
df_selected[df_selected['levelKBJI'] == 'Level KBJI Digit 4']
```

```python
len(df_selected)
```

```python
# Replace 'Level KBJI Digit 4' with NaN in 'levelKBJI' column
df_selected['levelKBJI'] = df_selected['levelKBJI'].replace('Level KBJI Digit 4', np.nan)
df_selected['levelKBJI'] = df_selected['levelKBJI'].replace('-', np.nan)
```

```python
df_selected.isna().sum()
```

```python


df_selected = df_selected.dropna(subset=['levelKBJI'])
df_selected = df_selected.dropna(subset=['jenjangKKNI'])
df_selected['levelKBJI Group'] = df_selected['levelKBJI'].astype(str).str[0].astype(int)
df_selected.head()
```

```python
df_selected.describe()
```

```python
cloudCols = ['instansi', 'program']
for i in cloudCols:
    text_data = ' '.join(df_selected[i].astype(str))

    # List of words to remove
    words_to_remove = ['DUDI', 'dan', 'Untuk', 'Menjadi', 'untuk', 'DAN']  # Add more words to remove as needed

    # Remove specified words from the text
    for word in words_to_remove:
        text_data = text_data.replace(word, '')

    # Generate the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)

    # Display the word cloud using Matplotlib
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off') 
```

```python
df_selected['jenjangKKNI'].value_counts().plot(kind='bar')
plt.title(f'Distribusi jenjangKKNI')
```
